```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/afnan47/sem7/raw/main/ML/6.%20KMeans%20on%20Sales/sales_data_sample.csv', encoding = 'latin-1')
```

```python
df.head()
```

```python
df.columns
```

```python
x = df[['QUANTITYORDERED','SALES']]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_scaled = sc.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss =[]
for i in range(1,11):
  km = KMeans(n_clusters = i)
  km.fit(x_scaled)
  wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
km = KMeans(n_clusters = 3)
```

```python
km.fit(x_scaled)
```
