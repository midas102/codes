```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x + 3) ** 2  # Corrected the power operator

def df(x):
    return 2 * (x + 3)  # Corrected the multiplication operator

def gradient_descent(initial_x, learning_rate, num_iterations):
    x = initial_x
    x_history = [x]  # Corrected the initialization of x_history

    for i in range(num_iterations):
        gradient = df(x)  # Corrected the assignment of gradient
        x -= learning_rate * gradient  # Corrected the update step
        x_history.append(x)

    return x, x_history

initial_x = 0  # Corrected the variable name
learning_rate = 0.1
num_iterations = 50  # Corrected the variable name

x, x_history = gradient_descent(initial_x, learning_rate, num_iterations)  # Corrected the assignment

print("Local minimum: {:.2f}".format(x))  # Corrected the print statement

# Create a range of x values to plot
x_vals = np.linspace(-10, 4, 100)  # Corrected the variable name

# Plot the function f(x)
plt.plot(x_vals, f(x_vals), label='f(x) = (x + 3)^2')

# Plot the values of x at each iteration
plt.plot(x_history, f(np.array(x_history)), 'r', label='Gradient Descent Steps')

# Label the axes and add a title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent')
plt.legend()  # Added a legend to differentiate the plots

# Show the plot
plt.grid()  # Optional: Add a grid for better readability
plt.show()
```
